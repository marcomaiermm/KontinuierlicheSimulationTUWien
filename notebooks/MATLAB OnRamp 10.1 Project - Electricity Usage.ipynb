```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Task Setup
data: np.array = np.random.uniform(low=2.5,high=5,size=(23,3))
# Set random Value to NaN
nan_rindex: int = np.random.choice(list(range(22)))
nan_cindex: int = np.random.choice(list(range(2)))
# Set Nan Value
data[nan_rindex,nan_cindex] = np.nan
data
```

```python
# Task 1 One of the elements in the usage variable has a value of NaN. Replace this value with the value 2.74.
data[np.isnan(data)] = 2.74
data
```

```python
# Task 2 The residential data is stored in the first column. Create a variable res that contains the first column of usage.
res: np.array = data[:,0]
res
```

```python
# Task 3 The commercial data and industrial data are stored in the second and third column, respectively.
# Create variables comm and ind that contain the second and third columns of usage.
comm: np.array = data[:,1]
ind: np.array = data[:,2]
(comm, ind)
```

```python
# Task 4/5 Create a vector named yrs that represents the years starting at 1991 and ending with 2013.
yrs: np.array = np.array(range(1991,2014))
yrs
```

```python
# Task 6
# Create a plot with all three columns. Use yrs as the x-data. Use this order and these plot specifications:
# res: blue (b) dashed line (--)
# comm: black (k) dotted line (:)
# ind: magenta (m) dash-dot line (-.)
plt.plot(yrs,res,"b--")
plt.plot(yrs,comm,"k:")
plt.plot(yrs,ind,"m-.")

# Task 7
# Add the title "July Electricity Usage" to the existing plot.
# Create a legend with the values "res", "comm", and "ind".
plt.title("July Electricity Usage")
plt.legend(["res","comm","ind"])
```
