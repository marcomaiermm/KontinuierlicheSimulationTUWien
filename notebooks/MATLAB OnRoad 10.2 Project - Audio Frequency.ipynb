```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Task Setup
y: np.array = np.random.uniform(low=-0.8454,high=0.596,size=12288)
fs: int = 8192
{"fs":fs,"y":y}
```

```python
# Task 1 Create a variable named n that contains the number of elements in y.
# Then use n to create an evenly-spaced vector t that starts at 0, ends at n-1, and has elements that are spaced by 1.
n: int = y.size
t: np.array = np.arange(0,n)
```

```python
# Task 2 Divide t by fs. Assign the output to the same variable t. Then plot y against t.
t = t/fs
plt.plot(t,y)
```

```python
# Task 3 Create a variable named yfft that contains the absolute value of the discrete Fourier transform of y.
yfft: np.array = np.abs(np.fft.fft(y))
yfft
```

```python
# Task 4 Create an evenly-spaced vector f that starts at 0, ends at n-1, and has elements that are spaced by 1.
f: np.array = np.arange(0,n)
```

```python
# Task 5
# Multiply f with fs/n. Assign the output to the same variable f.
# Plot yfft against f using the x-limits 0 and 1000.
f = f * fs/n
max_value: float = np.max(yfft)
plt.plot(f,yfft)
plt.xlim([0,1000])
plt.ylim([0,max_value])
```
