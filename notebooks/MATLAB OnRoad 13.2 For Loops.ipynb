```python
from typing import Tuple, Union, List
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def min_(data: np.ndarray) -> Tuple[Union[float,int]]:
    """
    Funktion simuliert das min() Verhalten aus Matlab
    """
    return (np.min(data),np.argmin(data))
```

```python
# SETUP
data: np.array = np.random.uniform(low=2.214e-12,high=8.584e-12,size=(357,7))
nObs: int = data.shape[0]
lambdaStart: float = 630.02
lambdaDelta: float = 0.14
    
lambdaEnd: float = lambdaStart + (nObs-1)*lambdaDelta
lambda_: np.ndarray = np.arange(start=lambdaStart, stop=lambdaEnd, step=lambdaDelta).T

s: np.ndarray = data[:,5]
```

```python
# TASK 1 Modify line 2 of the script. Delete (:,2) to calculate the speed of all the stars.
ha_: List[tuple] = [min_(x) for x in data.T]
sHa = np.array([x[0] for x in ha_])

lambdaHa_: np.ndarray = np.array([lambda_[x[1]] for x in ha_])
z: np.ndarray = lambdaHa_/656.28 - 1
speed: np.ndarray = z * 299792.458
```

```python
# TASK 2-5
# 2. Create a for loop with a loop index called v. The loop index should progress through all columns of spectra (1 to 7).
# In the loop body, extract the vth column of spectra to a variable named s.

# 3. Add an if statement to the for loop body. If speed(v) is less than or equal to 0, create a loglog plot of s against lambda using a dashed line (--).

# 4. Add an else statement. If speed(v) is greater than 0, create a loglog plot of s against lambda using a line width of 3.
# After the for loop is complete, enter hold off.

# 5. Add a legend to the plot using the array starnames.

starnames = np.array(["HD 30584","HD 11032","HD 64191","HD 5211", "HD 56030", "HD 94028","SAO102986"])

for v in range(7):
    s: np.ndarray = data[:,v]
    if speed[v] <= 0:
        plt.plot(lambda_, s, '--')
        plt.yscale("log")
    else:
        plt.plot(lambda_,s,linewidth=3)
        plt.yscale("log")
    plt.legend(starnames)
```
