```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Task Setup
data: np.array = np.random.rand(7,4)
sample: np.array = data[:,0]
density: np.array = data[:,1]
v1: np.array = data[:,2]
v2: np.array = data[:,3]
# Density Array und Sample Array sortieren f√ºr eine sinnvollere Plot Darstellung
sample.sort()
density.sort()
mass1: np.array = density*v1
mass2: np.array = density*v2
```

```python
# Task 1 Create a plot with sample on the x-axis and mass1 on the y-axis.
# Sort density array
plt.plot(sample,mass1)
```

```python
# Task 2 Plot mass2 (y-axis) against sample (x-axis). Use red (r) star (*) markers and no line in your plot.
plt.plot(sample,mass2,"r*")
```

```python
# Task 3 Enter the hold on command.
# Then plot mass1 (y-axis) against sample (x-axis) with black (k) square (s) markers and no line.
plt.plot(sample,mass2,"r*")
plt.plot(sample, mass1, "ks")
```
